#### Ejercicio 

Estudiar el comportamiento de las trayectorias al variar $\mu$ y $\sigma$.

```python
from random import gauss
from math import exp
import matplotlib.pyplot as plt

mu=0.001 
sigma=0.04  #sigma
p0=10
t=30

def trayectoria(mu,sigma,p0,t):
    l=[p0]
    pt=p0
    for i in range(t):
        pt=pt*exp(mu + sigma*gauss(0,1))
        l.append(pt)
    return l

l=trayectoria(mu, sigma,p0,t)
print(l)

plt.plot(l)
plt.show()
```

```python
mu=0.001
sigma=0.04
p0=10
t=200

for i in range(10):
    l=trayectoria(mu,sigma,p0,t)
    plt.plot(l)
plt.show()
```

Con $\mu$ y $\sigma$ mayores

```python
mu=0.05
sigma=0.08
p0=10
t=200

for i in range(10):
    l=trayectoria(mu,sigma,p0,t)
    plt.plot(l)
plt.show()
```

Con $\mu$ menor 

```python
mu=0.00085
sigma=0.08
p0=10
t=200

for i in range(10):
    l=trayectoria(mu,sigma,p0,t)
    plt.plot(l)
plt.show()
```

Con  $\sigma$  menor

```python
mu=0.05
sigma=0.000098
p0=10
t=200

for i in range(10):
    l=trayectoria(mu,sigma,p0,t)
    plt.plot(l)
plt.show()
```

Con $\mu$ y $\sigma$ menores

```python
mu=0.000000065
sigma=0.00000098
p0=10
t=200

for i in range(10):
    l=trayectoria(mu,sigma,p0,t)
    plt.plot(l)
plt.show()
```
