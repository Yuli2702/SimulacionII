```python
from math import log
from random import random
import matplotlib.pyplot as plt

T=30       #Tiempo de simulacion 
l_l=2      #Usuarios por minuto
l_s=2.2    #Capacidad de servicio por minuto
t=0        #inicio de  la simulacion en t=0
infi=10000  #el primer tiempo de salida 
ts=infi    
tl=0       #Tiempo de llegada de los usuarios
l=0        #Numero de usuarios en el sistema
lt=[ ]     #Tiempos de llegada 
ll=[ ]     #Llegadas y salidas
while min(tl,ts)<T:
    lt.append(t)
    ll.append(l)
    t=min(tl,ts)
    lt.append(t)
    ll.append(l)
    if tl<=ts:
        l=l+1
        print("%10.2f"%tl,"          ",l)
        tl=t+(-1/l_l)*log(random())
        if l==1:ts=t+(-1/l_s)*log(random())
    else:
        l=l-1
        print("                                          ","%10.2f"%ts,"         ",l)
        if l==0:
            ts=infi
        else:
            ts=t+(-(1./l_s)*log(random()))
plt.plot(lt,ll,color="blue")
plt.xlabel("Tiempo")
plt.ylabel("Personas en el sistema")
plt.show()
```
