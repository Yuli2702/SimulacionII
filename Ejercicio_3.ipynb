```python
import random  
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1, 100)
y = np.sqrt(1 - x**2)     

x_1 = random.random()
y_1 = random.random()

fig = plt.figure(1, figsize=(5, 5))     
plt.plot(x, y, color="pink", markersize=1) 
plt.scatter(x_1, y_1, s=20, c="green")
plt.title("Estimación de $\pi$")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.show()
```

```python
def pi_Taylor(n):
    pi = 0
    L =[]
    for i in range(n):
        pi = pi + 4.0/(2*i+1)*(-1)**i
        L.append(pi)
    fig = plt.figure(1, figsize=(8, 8))        
    plt.plot(L, color="red", markersize=1)
    plt.xlabel("Iteraciones")
    plt.ylabel("Estimación")
    plt.grid()
    plt.show()
```

```python
pi_Taylor(2000)
```
