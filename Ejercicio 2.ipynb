b) Simular 5 sorteos con sustitución.

```python
import random 
def sorteo():

    x=random.randint(1,5)
    if x<=1:
        color = 'blanco'
    elif x<=2:
        color = 'negro'
    elif x<=3:
        color = 'rojo'
    elif x<=4:
        color = 'azul'
    else:
        color = 'verde'
    return color
print(sorteo())
```
