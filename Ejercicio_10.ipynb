# **triangular()**

```python
import random 
import matplotlib.pyplot as plt 
nums = [] 
low = 0.0 
high = 1.0
mode=None 
for i in range(100): 
  temp = random.triangular(low,high,mode) 
  nums.append(temp) 
plt.title("Números aleatorios")
plt.plot(nums) 
plt.show() 
```

```python
import random 
import matplotlib.pyplot as plt 
nums = [] 
low = 0.0 
high = 1.0
mode=None 
for i in range(1000): 
  temp = random.triangular(low,high,mode) 
  nums.append(temp) 
plt.title("Histograma")
plt.hist(nums, bins = 200) 
plt.show()  
```
