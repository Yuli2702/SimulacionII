Simular dos distribuciones de probabilidad (no la normal, no la exponencial), por el método que prefieran.

**Distribución de Poisson**

```python
import matplotlib.pyplot as plt
import numpy as np 
from scipy import stats 
import seaborn as sns 

mu =  50
poisson = stats.poisson(mu) # Distribución
x = np.arange(poisson.ppf(0.01), poisson.ppf(0.99))
fmp = poisson.pmf(x)
aleatorios = poisson.rvs(10000)  
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Distribución Poisson')
plt.show()
```

**Distribución geométrica**

```python
import matplotlib.pyplot as plt
import numpy as np 
from scipy import stats 
import seaborn as sns 

p =  0.3
geometrica = stats.geom(p) 
x = np.arange(geometrica.ppf(0.01), geometrica.ppf(0.99))
fmp = geometrica.pmf(x) 

aleatorios = geometrica.rvs(10000)  
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Distribución Geométrica')
plt.show()
```
